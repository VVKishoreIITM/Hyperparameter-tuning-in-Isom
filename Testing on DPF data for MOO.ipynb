```python
def classify_all_columns(df):
    column_types = {}
    for column in df.columns:
        unique_values = df[column].nunique()
        if df[column].dtype == 'object' or df[column].dtype.name == 'category' or unique_values <= 10:
            if unique_values == 2:
                column_types[column] = 'Binary Classifier'
            elif unique_values == 3:
                column_types[column] = 'Tertiary Classifier'
            else:
                column_types[column] = 'Other Classifier'
        else:
            column_types[column] = 'Continuous'
    return column_types

# Classify all columns in the DataFrame
classified_columns = classify_all_columns(df)

# Print classification of all columns
for column, column_type in classified_columns.items():
    print(f"{column}: {column_type}")
```

```python
som_models = []  # List to store all trained SOM models

for i in range(1, outputs+1):
    # Get n_array and cnames dynamically
    n_array = globals()[f"n_array{i}"]
    cnames = globals()[f"cnames{i}"]

    # Build the SOM model
    som = smf.build(
        n_array,
        normalization='range',
        initialization='pca',
        mapshape="planar",
        lattice="hexa",
        neighborhood='gaussian',
        training='batch',
        mapsize=sm0.codebook.mapsize,
        component_names=cnames
    )
    
    # Set initialized matrix
    som.set_initialized_matrix(sm0.initialized_matrix)
    
    # Train the SOM model
    som.train(
        request_id=None,
        n_job=1,
        shared_memory=False,
        verbose='info',
        train_rough_len=500,
        train_finetune_len=500,
        train_len_factor=1
    )
    
    # Append trained model to the list
    som_models.append(som)
    
for i in range(1,  outputs+1 ):
    globals()[f"sm{i}"] = som_models[i-1]
```
